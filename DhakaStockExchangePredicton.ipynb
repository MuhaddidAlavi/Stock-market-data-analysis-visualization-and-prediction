```
from sklearn.preprocessing import RobustScaler
f_columns = ['month','day_of_month','day_of_year','day_of_week','LTP*', 'HIGH', 'LOW','OPENP*','difference',
             'CLOSEP*','YCP','TRADE','VALUE (mn)']
f_transformer = RobustScaler()
f_transformer = f_transformer.fit(train[f_columns].to_numpy())
train.loc[:, f_columns] = f_transformer.transform(train[f_columns].to_numpy())
test.loc[:, f_columns] = f_transformer.transform(test[f_columns].to_numpy())
```

```
cnt_transformer = RobustScaler()
cnt_transformer = cnt_transformer.fit(train[['CLOSEP*']])
train['CLOSEP*'] = cnt_transformer.transform(train[['CLOSEP*']])
test['CLOSEP*'] = cnt_transformer.transform(test[['CLOSEP*']])
```

```
def create_dataset(X, y, time_steps=1):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        v = X.iloc[i:(i + time_steps)].values
        Xs.append(v)
        ys.append(y.iloc[i + time_steps])
    return np.array(Xs), np.array(ys)
```

```
TIME_STEPS = 10
#reshape to [samples, time_steps, n_features]
X_train, y_train = create_dataset(train, train.YCP,time_steps=TIME_STEPS)
X_test, y_test = create_dataset(test, test.YCP, time_steps=TIME_STEPS)
print(X_train.shape, y_train.shape)
```

```
print(X_test.shape, y_test.shape)
```

```
X_test[0][0]
```

```
X_test[0][0].shape
```

```
model = keras.Sequential()
model.add(
  keras.layers.Bidirectional(
    keras.layers.LSTM(
      units=128,
      input_shape=(X_train.shape[1], X_train.shape[2])
    )
  )
)
#model.add(keras.layers.Dropout(rate=0.2))
model.add(keras.layers.Dense(units=1))
model.compile(loss='mean_squared_error', optimizer='adam')
```

```
history = model.fit(X_train, y_train,epochs=30,batch_size=32,validation_split=0.1,shuffle=False)
```

```
plt.plot(history.history['loss'],label='train')
plt.plot(history.history['val_loss'],label='validation')
plt.legend()
```

```
y_pred = model.predict(X_test)
```
